```python
#https://github.com/chekos/altair_themes_blog/blob/master/notebooks/vox_theme.py

def vox_theme():
    markColor = '#3e5c69'
    
    return {
        'config': {
            'background': '#fff',
            'arc': { 
                'fill': markColor 
            },
            'area': { 
                'fill': markColor 
            },
            'line': { 
                'stroke': markColor 
            },
            'path': { 
                'stroke': markColor 
            },
            'rect': { 
                'fill': markColor 
            },
            'shape': { 
                'stroke': markColor 
            },
            'symbol': { 
                'fill': markColor 
            },
            'axis': {
                'domainWidth': 0.5,
                'grid': True,
                'labelPadding': 2,
                'tickSize': 5,
                'tickWidth': 0.5,
                'titleFontWeight': 'normal',
            },
            'axisBand': {
                'grid': False,
            },
            'axisX': {
                'gridWidth': 0.2,
            },  
            'axisY': {
                'gridDash': [3],
                'gridWidth': 0.4,
            },
            'legend': {
                'labelFontSize': 11,
                'padding': 1,
                'symbolType': 'square',
            },
            'range': {
                'category': [
                    '#30a2da',
                    '#fc4f30',
                    '#2000ee',
                    '#6d904f',
                    '#8b8b8b',
                    '#b96db8',
                    '#ff9e27',
                    '#56cc60',
                    '#52d2ca',
                    '#52689e',
                    '#545454',
                    '#9fe4f8',
                ],
            },
        }
        
    }

import altair as alt

# register the custom theme under a chosen name
alt.themes.register('vox_theme', vox_theme)

# enable the newly registered theme
alt.themes.enable('vox_theme')
```

# Getting Data:

```python
df = {}

df['S&P'] = fred.get_series('SP500') # Daily Close
# df['gdp'] = fred.get_series('GDP') # Gross Domestic Product, SA
# df['M2SL'] = fred.get_series('M2SL') # M2 (money supply), SA 
df['DFF'] = fred.get_series('DFF') # Federal Funds Effective Rate (DFF), daily
df['DGS3MO'] = fred.get_series('DGS3MO') # 3 month treasury yield
df['DGS6MO'] = fred.get_series('DGS6MO') # 6 month treasury yield
df['DGS10'] = fred.get_series('DGS10') # 10 yr treasury yield
df['BAMLH0A0HYM2'] = fred.get_series('BAMLH0A0HYM2') # tracks the performance of US dollar denominated below investment grade rated corporate debt publicly issued in the US domestic market.
# df['CPI'] = fred.get_series('CPALTT01USM661S') # Consumer Price Index: Total All Items for the United States, SA
# df['PPIACO'] = fred.get_series('CPALTT01USM661S') # PPI all commodities, NOT SA !!
# df['PAYEMS'] = fred.get_series('PAYEMS') # All Employees, Total Nonfarm, SA
# df['UNRATE'] = fred.get_series('UNRATE') # Unemployment rate, SA
# df['UMCSENT'] =fred.get_series('UMCSENT') # UMICH Consumer Sentiment, NOT SA!!
df['VXIBMCLS'] =fred.get_series('VXIBMCLS') # CBOE VIX (daily)
df['USEPUINDXD'] = fred.get_series('USEPUINDXD') # The daily news-based Economic Policy Uncertainty Index is based on newspapers in the United States.
df['DCOILWTICO'] = fred.get_series('DCOILWTICO') # Crude Oil Prices: West Texas Intermediate (WTI), daily


df = pd.DataFrame(df)
df = df.rename_axis('dt').reset_index()
```

# EDA : Missing value - count and percentage

```python
print('Null Values Present: ', df.isnull().values.sum())
```

```python
count_missing = df.isnull().sum()
percent_missing = df.isnull().sum() * 100 / len(df)

missing_value_df = pd.DataFrame(
    {'count_missing': count_missing,
     'percent_missing': percent_missing}
)
missing_value_df
```

```python
# forward fill to impute missing data, drop NA to drop dates where records didn't start.

df_ff = df.ffill().dropna() 
df_ff
```

```python
count_missing = df_ff.isnull().sum()
percent_missing = df_ff.isnull().sum() * 100 / len(df_ff)

missing_value_df = pd.DataFrame(
    {'count_missing': count_missing,
     'percent_missing': percent_missing}
)
missing_value_df
```
