# Import Price/Technical Dataset

```python
start = datetime.datetime(2012, 1, 1) # START ON January 1 2017
end = datetime.date.today() # LOOK AT STOCKS TILL current date

tickers = ['KO','PEP']
df_map = {}

for tick in tickers:
    data_source='google'
    stock =yf.download(tick,start, end)
    df_map[tick] = stock
```

```python
KO_PEP = pd.concat(df_map,axis = 1,keys = tickers)
# KO_PEP.columns.names = ['Stock Ticker','Stock Info']
# close_price = KO_PEP.xs(key='Close',axis=1,level='Stock Info')

KO_PEP.columns = ['KO_Open','KO_High','KO_Low','KO_Close','KO_AdjClose','KO_Volume',
                  'PEP_Open','PEP_High','PEP_Low','PEP_Close','PEP_AdjClose','PEP_Volume']
KO_PEP
```

## Stochastic Oscillator (%K) - warns of strength or weakness of market

```python
#  https://www.learnpythonwithrune.org/pandas-calculate-the-stochastic-oscillator-indicator-for-stocks/
#  https://www.investopedia.com/terms/s/stochasticoscillator.asp

KO_PEP['KO_14-high'] = KO_PEP['KO_High'].rolling(14).max()
KO_PEP['KO_14-low'] = KO_PEP['KO_Low'].rolling(14).min()
KO_PEP['KO_%K'] = (KO_PEP['KO_Close'] - KO_PEP['KO_14-low'])*100/(KO_PEP['KO_14-high'] - KO_PEP['KO_14-low'])
KO_PEP['KO_%D'] = KO_PEP['KO_%K'].rolling(3).mean()

KO_PEP['PEP_14-high'] = KO_PEP['PEP_High'].rolling(14).max()
KO_PEP['PEP_14-low'] = KO_PEP['PEP_Low'].rolling(14).min()
KO_PEP['PEP_%K'] = (KO_PEP['PEP_Close'] - KO_PEP['PEP_14-low'])*100/(KO_PEP['PEP_14-high'] - KO_PEP['PEP_14-low'])
KO_PEP['PEP_%D'] = KO_PEP['PEP_%K'].rolling(3).mean()
KO_PEP
```

## Larry William (%R) - indicator to spot overbought/oversold levels

```python
# https://www.investopedia.com/terms/w/williamsr.asp
# https://medium.com/codex/algorithmic-trading-with-williams-r-in-python-5a8e0db9ff1f

KO_PEP['KO_%R'] = -100 * ((KO_PEP['KO_14-high'] - KO_PEP['KO_Close']) / (KO_PEP['KO_14-high'] - KO_PEP['KO_14-low']))
KO_PEP['PEP_%R'] = -100 * ((KO_PEP['PEP_14-high'] - KO_PEP['PEP_Close']) / (KO_PEP['PEP_14-high'] - KO_PEP['PEP_14-low']))
```

# RSI 

```python
# https://www.roelpeters.be/many-ways-to-calculate-the-rsi-in-python-pandas/

def rsi(df, column, periods = 14, ema = True):
    """
    Returns a pd.Series with the relative strength index.
    """
    close_delta = df[column].diff()

    # Make two series: one for lower closes and one for higher closes
    up = close_delta.clip(lower=0)
    down = -1 * close_delta.clip(upper=0)
    
    if ema == True:
	    # Use exponential moving average
        ma_up = up.ewm(com = periods - 1, adjust=True, min_periods = periods).mean()
        ma_down = down.ewm(com = periods - 1, adjust=True, min_periods = periods).mean()
    else:
        # Use simple moving average
        ma_up = up.rolling(window = periods, adjust=False).mean()
        ma_down = down.rolling(window = periods, adjust=False).mean()
        
    rsi = ma_up / ma_down
    rsi = 100 - (100/(1 + rsi))
    return rsi

# KO_PEP['KO_Close'] = KO_PEP['KO']['Close']
# KO_PEP['PEP_Close'] = KO_PEP['PEP']['Close']
KO_PEP['KO_RSI'] = rsi(df=KO_PEP, column='KO_Close')
KO_PEP['PEP_RSI'] = rsi(df=KO_PEP, column='PEP_Close')
```
