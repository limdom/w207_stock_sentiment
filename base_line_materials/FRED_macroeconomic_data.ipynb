```python
#https://github.com/chekos/altair_themes_blog/blob/master/notebooks/vox_theme.py

def vox_theme():
    markColor = '#3e5c69'
    
    return {
        'config': {
            'background': '#fff',
            'arc': { 
                'fill': markColor 
            },
            'area': { 
                'fill': markColor 
            },
            'line': { 
                'stroke': markColor 
            },
            'path': { 
                'stroke': markColor 
            },
            'rect': { 
                'fill': markColor 
            },
            'shape': { 
                'stroke': markColor 
            },
            'symbol': { 
                'fill': markColor 
            },
            'axis': {
                'domainWidth': 0.5,
                'grid': True,
                'labelPadding': 2,
                'tickSize': 5,
                'tickWidth': 0.5,
                'titleFontWeight': 'normal',
            },
            'axisBand': {
                'grid': False,
            },
            'axisX': {
                'gridWidth': 0.2,
            },  
            'axisY': {
                'gridDash': [3],
                'gridWidth': 0.4,
            },
            'legend': {
                'labelFontSize': 11,
                'padding': 1,
                'symbolType': 'square',
            },
            'range': {
                'category': [
                    '#30a2da',
                    '#fc4f30',
                    '#2000ee',
                    '#6d904f',
                    '#8b8b8b',
                    '#b96db8',
                    '#ff9e27',
                    '#56cc60',
                    '#52d2ca',
                    '#52689e',
                    '#545454',
                    '#9fe4f8',
                ],
            },
        }
        
    }

import altair as alt

# register the custom theme under a chosen name
alt.themes.register('vox_theme', vox_theme)

# enable the newly registered theme
alt.themes.enable('vox_theme')
```

# Getting Data:

```python
# not in use..

def fred_to_df(fred_id: str) -> pd.DataFrame:
    """ Helper function to convert API call from Series to Dataframe.

    param(s):
    -------
    fred_id: FRED Economic Indicator ID, i.e  PAYEMS = All Employees, Total Nonfarm
    
    return: pd.DataFrame
    """
    return fred.get_series(str(fred_id)).to_frame()
```

```python
df = {}

df['S&P'] = fred.get_series('SP500') # Daily Close
df['gdp'] = fred.get_series('GDP') # Gross Domestic Product, SA
df['M2SL'] = fred.get_series('M2SL') # M2 (money supply), SA
df['FEDFUNDS'] = fred.get_series('FEDFUNDS') # Federal Funds Effective Rate
df['DGS3MO'] = fred.get_series('DGS3MO') # 3 month treasury yield
df['DGS6MO'] = fred.get_series('DGS6MO') # 6 month treasury yield
df['DGS10'] = fred.get_series('DGS10') # 10 yr treasury yield
df['BAMLH0A0HYM2'] = fred.get_series('BAMLH0A0HYM2') # tracks the performance of US dollar denominated below investment grade rated corporate debt publicly issued in the US domestic market.
df['CPI'] = fred.get_series('CPALTT01USM661S') # Consumer Price Index: Total All Items for the United States, SA
df['PPIACO'] = fred.get_series('CPALTT01USM661S') # PPI all commodities, NOT SA !!
df['PAYEMS'] = fred.get_series('PAYEMS') # All Employees, Total Nonfarm, SA
df['UNRATE'] = fred.get_series('UNRATE') # Unemployment rate, SA
df['UMCSENT'] =fred.get_series('UMCSENT') # UMICH Consumer Sentiment, NOT SA!!

df = pd.DataFrame(df)
df = df.rename_axis('dt').reset_index()
```

# Charting for baseline presentation:

```python
df_markets = df.set_index('dt').loc['2019-01-01':'2022-07-01']
df_markets['CPI_%_CHANGE'] = df_markets['CPI'].pct_change(freq='1y')*100
df_markets['FEDFUNDS_%_CHANGE'] = df_markets['FEDFUNDS'].pct_change(freq='1d')*100
df_markets['S&P_%_CHANGE'] = df_markets['S&P'].pct_change(freq='30d')*100
df_markets['M2SL_%_CHANGE'] = df_markets['M2SL'].pct_change(freq='1m')*100

df_markets = df_markets.ffill().dropna()
df_markets
```

```python
alt.Chart(df_markets.reset_index()).mark_line().encode(
    x=alt.X("dt", title='Date'),
    y=alt.Y(alt.repeat('layer'), aggregate='mean', title="Percent (%)"),
    color=alt.ColorDatum(alt.repeat('layer'))
).properties(
    width=800, height=400).repeat(layer=["S&P_%_CHANGE","CPI_%_CHANGE","UNRATE","FEDFUNDS"])
```

```python
# needs work.. need to specify percentage change by year/month/quarter.

def add_pct_change(dataframe: pd.DataFrame) -> pd.DataFrame:
    for column in dataframe.columns:
        if column == 'dt':
            continue
        else:
            dataframe[column+'pct_change'] = dataframe[column].pct_change()
    return dataframe
```
