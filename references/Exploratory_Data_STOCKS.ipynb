```python
start = datetime.datetime(2012, 1, 1) # START ON January 1 2017
end = datetime.date.today() # LOOK AT STOCKS TILL current date

tickers = ['MSFT','AAPL','AMZN','GOOG','BABA','FB','INTC','NVDA','CRM','PYPL','TSLA','AMD','EA','MTCH','TTD','ZG','ATVI','YELP','ZM','VMW']
df_map = {}

for tick in tickers:
    data_source='google'
    stock =yf.download(tick,start, end)
    df_map[tick] = stock
```

```python
tech_stocks = pd.concat(df_map,axis = 1,keys = tickers)
tech_stocks.columns.names = ['Stock Ticker','Stock Info']
tech_stocks.head()
```

```python
tech_stocks.to_csv('data.csv')
```

```python
aapl_df = yf.download('AAPL', 
                      start='2019-01-01', 
                      end='2022-02-01', 
                      progress=False,
)
aapl_df.tail()
```

```python
## CLOSING PRICE
price = alt.Chart(aapl_df.reset_index()).mark_line().encode(
    alt.X("Date", title = "Date"),
    alt.Y("Close", title = "Closing"),
    tooltip= ['Date', 'Open', 'Close', 'High', 'Low', 'Volume']
).properties(
    width=600, height=200
)

price
```
